# KNN

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```python
from sklearn.model_selection import GridSearchCV

def grid_search(estimator, target, scoring, nn=30, n_iter=50, cv=10, n_jobs=-1, n_top=1, verbose=True):
    
    param_grid = {
    "imputer__strategy": ["most_frequent", "mean", "median"],
    "pca__n_components": [0.6, 0.7, 0.8, 0.85, 0.9, 0.95, 0.99],
    "knn__n_neighbors": list(range(1,nn))
    "knn__weights": ["uniform", "distance"],
    "knn__metric": ["euclidean", "manhattan"]
    }
    
    imputer = SimpleImputer()
    scaler = StandardScaler()
    pca = PCA()
    pipe = Pipeline(steps=[('imputer', imputer), ('scaler', scaler), ('pca', pca), ('knn', estimator)])
    
    # train using only labelled data
    target_not_null = np.logical_not(y_train.loc[:, target].isnull())
    
    # run randomised search
    random_search = GridSearchCV(pipe, param_grid=param_grid, cv=cv, n_jobs=n_jobs,
                                       scoring=scoring)
    random_search.fit(X_train[target_not_null], y_train[target_not_null].loc[:, target])
    
    if verbose:
        report(random_search.cv_results_, n_top=n_top)
    
    return random_search.best_estimator_
```

# Flourishing

```python
knn_gs_flour_class_pre = grid_search(KNeighborsClassifier(algorithm="auto"),
                                     "flourishing_scale_imp_class_pre", "neg_log_loss")
```

```python
knn_gs_flour_pre = grid_search(KNeighborsClassifier(algorithm="auto"),
                               "flourishing_scale_imp_pre", "neg_mean_squared_error", cv=4, nn=15)
```
